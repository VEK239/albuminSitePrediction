```python
import pandas as pd
```

```python
broken_data = pd.read_csv('../data/lexa_broken_names.csv')
old_data = pd.read_csv('../data/binding_data.csv')
data = pd.read_csv('../data/lexa_preprocessed.csv')
```

```python
old_data.head()
```

```python
broken_data
```

```python
broken_data = broken_data.dropna()
```

```python
old_name_to_binding = {k: bool(v) for k, v in zip(old_data.ligand, old_data['Binder/\nNonbinder'])}
```

```python
broken_data.columns
```

```python
active = pd.Series([int(not old_name_to_binding[x]) for x in broken_data['Old name '].values])
```

```python
broken_data = broken_data.reset_index()
```

```python
broken_data = broken_data[['smiles']]
```

```python
broken_data['active'] = active
```

```python
broken_data
```

```python
new_data = pd.concat([data, broken_data])
```

```python
new_data.to_csv('../data/lexa')
```
