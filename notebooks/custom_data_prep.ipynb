## QikProp

```python
qikprop_data = pd.read_csv('../data/qikprop/custom_test_data.csv')
```

```python
qikprop_data
```

```python
qikprop_data = qikprop_data.drop(columns=['row ID', 'CT_input', 'CT', '#stars'])
```

```python
qikprop_data
```

```python
data = pd.read_csv('../data/qikprop/qikprop_data.csv')
```

```python
data = data.drop(columns=['row ID', 'CT_input', 'CT', '#stars', 'index', 'active'])
```

```python
data = pd.concat([data, qikprop_data])
```

```python
dict_data = data.to_dict(orient='records')
```

```python
smiles_to_features = {}
for dct in dict_data:
    smiles = dct['smiles']
    del dct['smiles']
    smiles_to_features[smiles] = dct
```

```python
with open('../data/qikprop/qikprop.json', 'w') as ouf:
    json.dump(smiles_to_features, ouf)
```
